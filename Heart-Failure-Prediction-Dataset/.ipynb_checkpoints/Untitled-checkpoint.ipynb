```python
# Probably no need to under or oversample because the Ratio of healthy and with disease is = ~ 1
data["HeartDisease"].value_counts()
```

```python
# used Robust scaler to scale for interquartile range

scaler = preprocessing.RobustScaler()
robust_df = scaler.fit_transform(data.select_dtypes(np.integer))
robust_df = pd.DataFrame(robust_df, columns=data.select_dtypes(np.integer).columns )
```
